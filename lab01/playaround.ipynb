# Function that just display things

```python
# Import required libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os
from pathlib import Path
# start play



image_dir = 'COMP9517_26T2_Lab1_Images'

# path create it can be used dynamicly
filepath = os.path.join(image_dir, 'Sydney_01.jpg')
# by default it is a BGR you want to display it
# open image by cv
img = cv.imread(filepath)
print(type(img))
# 这里竟然是一个np araay

img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
print(type(img_rgb.shape))
print(img_rgb.shape)
print(img.shape)


#plt.axis('off')  # 贴心小技巧：关掉坐标轴刻度线，让作业界面更干净美观
plt.show()




```
